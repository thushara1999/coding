# Dealing with outliers

```python
upper_limit = data['loudness'].mean() + 1.5*data['loudness'].std()
lower_limit = data['loudness'].mean() - 1.5*data['loudness'].std()

data['loudness'] = np.where(
    data['loudness']>upper_limit,
    upper_limit,
    np.where(
        data['loudness']<lower_limit,
        lower_limit,
        data['loudness']))
```

```python
upper_limit = data['speechiness'].mean() + 3*data['speechiness'].std()
lower_limit = data['speechiness'].mean() - 3*data['speechiness'].std()

data['speechiness'] = np.where(
    data['speechiness']>upper_limit,
    upper_limit,
    np.where(
        data['speechiness']<lower_limit,
        lower_limit,
        data['speechiness']
    )
)
```

# Model building

```python
X_Data=data.iloc[:,0:9]
Y_Data=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X_Data,Y_Data,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model1 = GaussianNB()
model1.fit(X_train,Y_train)
```

```python
y_predicted = model1.predict(X_test)
```

```python
from sklearn import metrics
 
metrics.accuracy_score(y_predicted , Y_test)
```
